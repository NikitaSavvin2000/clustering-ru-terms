```python
import pandas as pd

def get_indices_between_dates(df_Pl_sacler, start_date, end_date):
    """
    Получает DataFrame, начальную и конечную даты, возвращает список индексов,
    соответствующих датам в заданном диапазоне.
    Нужна для получения индексов времени которые мы хотим спрогнозировать
    Parameters:
    - df_Pl_sacler: pandas.DataFrame
        DataFrame с колонкой 'time', содержащей даты в формате 'yyyy-mm-dd hh:mm:ss+00:00'.
    - start_date: str
        Начальная дата в формате 'yyyy-mm-dd'.
    - end_date: str
        Конечная дата в формате 'yyyy-mm-dd'.

    Returns:
    - list
        Список индексов, соответствующих датам в заданном диапазоне.
    """
    # Преобразование колонки 'time' в формат datetime
    df_Pl_sacler['pubdate'] = pd.to_datetime(df_Pl_sacler['pubdate'])

    # Выборка индексов для дат в заданном диапазоне
    indices = df_Pl_sacler.loc[(df_Pl_sacler['pubdate'] >= start_date) & (df_Pl_sacler['pubdate'] <= end_date)].index

    return indices

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error


def calculate_metrics(df_orig, df_test):
    avg_pl = df_Pl['count'].mean()

# Заполняем Nan если они есть в датафрейме с известными данными
    df_orig['count'].fillna(avg_pl, inplace=True)


    y_true = df_orig
    y_pred = df_test

    y_true_mean = y_true['count'].mean()
        # Расчет RMSE
    rmse = np.sqrt(mean_squared_error(y_true['count'].values , y_pred['count'].values ))

    # Расчет R^2
    ss_res = np.sum((y_true['count'].values  - y_pred['count'].values ) ** 2)
    ss_tot = np.sum((y_true['count'].values  - y_true_mean) ** 2)
    r2 = 1 - (ss_res / ss_tot)

    # Расчет MAE
    mae = mean_absolute_error(y_true['count'].values , y_pred['count'].values )

    # Расчет MAPE
    mape = np.mean(np.abs((y_true['count'].values - y_pred['count'].values) / y_true['count'].values)) * 100
    # Расчет WMAPE
    wmape = np.sum(np.abs(y_true['count'].values  - y_pred['count'].values )) / np.sum(np.abs(y_true['count'].values )) * 100

    return rmse, r2, mae, mape, wmape
```
